```python
# Write a program to perform rotation, scaling, flipping and resizing on an image.

%pip install pillow
from PIL import Image
```

```python
# Rotation


def rotate_img_90_clockwise(imgPath):
    img = Image.open(imgPath)
    return img.rotate(-90, expand=True)  # Negative for clockwise


def rotate_img_90_counter_clockwise(imgPath):
    img = Image.open(imgPath)
    return img.rotate(90, expand=True)  # Positive for counter-clockwise


def rotate_img_180(imgPath):
    img = Image.open(imgPath)
    return img.rotate(180, expand=True)


# Flipping


def flip_img_vertically(imgPath):
    img = Image.open(imgPath)
    return img.transpose(Image.Transpose.FLIP_TOP_BOTTOM)


def flip_img_horizontally(imgPath):
    img = Image.open(imgPath)
    return img.transpose(Image.Transpose.FLIP_LEFT_RIGHT)


def flip_img_horizontally_vertically(imgPath):
    img = Image.open(imgPath)
    return img.transpose(Image.Transpose.FLIP_LEFT_RIGHT).transpose(Image.Transpose.FLIP_TOP_BOTTOM)


# scalling


def scale_img(imgPath, scale_factor):
    img = Image.open(imgPath)
    new_size = (int(img.width * scale_factor), int(img.height * scale_factor))
    return img.resize(new_size)


# Resizing
def resize_img(imgPath, new_size):
    img = Image.open(imgPath)
    return img.resize(new_size)
```

```python
rotate_img_90_clockwise("image.jpg")
```

```python
rotate_img_90_counter_clockwise("image.jpg")
```

```python
rotate_img_180("image.jpg")
```

```python
flip_img_vertically("image.jpg")
```

```python
flip_img_horizontally("image.jpg")
```

```python
flip_img_horizontally_vertically("image.jpg")
```

```python
scale_img("image.jpg", 1.5)
```

```python
resize_img("image.jpg", (200, 200))
```
